```python
#importación de bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np
```

```python
#Cargar el copnjunto de datos Iris
iris = sns.load_dataset('iris')
```

```python
#Mostrar las primeras filas del conjunto de datos
print (iris.head())
```

```python
#Número de observaciones por categoría
print("\nNumero de Observaciones por especie:\n", iris['species'].value_counts())
```

```python
#Estadisticas Descriptivas
print("\nEstadisticas Descriptivas:\n", iris.describe())
```

```python
#histogramas para cada especie y cada variable
features = iris.columns[:-1] #todas las columnas excepto 'species'
for feature in features:
    g = sns.FacetGrid(iris, col='species')
    g.map(plt.hist, feature)
    plt.suptitle(f"Histograma de {feature} por especie", y=1.05)
    plt.show()
```

```python
#Scatter plots para cada par de caracteristicas
sns.pairplot(iris,hue='species')
plt.suptitle("scatter plots para cada par de caracteristicas", y=1.02)
plt.show()
```

```python
#Matriz de correlacion
# Seleccionar solo las columnas numéricas del conjunto de datos iris
numeric_columns = iris.select_dtypes(include=np.number)

# Calcular la matriz de correlación utilizando las columnas numéricas seleccionadas
correlation_matrix = round(numeric_columns.corr(),2)
#Old value 
#correlation_matrix = iris.corr()
print("\nMatrix de Correlacion: \n", correlation_matrix)
```

```python
#Visualizacion de la matriz de correlacion con heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap de correlaciones")
plt.show()
```

```python
#Boxplots para cada caracteristica por especie
for feature in features:
    plt.figure(figsize=(8,6))
    sns.boxplot(x="species",y=feature, data=iris)
    plt.title(f"Boxplot de {feature} por especie")
    plt.show()
```
